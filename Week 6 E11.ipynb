```python
plt.figure(figsize=(12,6))
sns.distplot(indoors['Temperature'],bins=20,label='Experimental Data')
plt.xticks(rotation=90);
print("The mean of the temperatures is {}.".format(np.mean(indoors['Temperature'])))
print("The sd of the temperatures is {}.".format(np.std(indoors['Temperature'])))

mu=np.mean(indoors['Temperature'])
sigma =np.std(indoors['Temperature'])
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
plt.plot(x, stats.norm.pdf(x, mu, sigma),label='Gaussian Distribution')

plt.legend(loc='upper right');

print("This is a multimodal distribution centered on three values: 22.1, 22.15, and 22.25 Celsius")
```
