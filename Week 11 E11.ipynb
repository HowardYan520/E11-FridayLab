## <font color='Green'> Histograms for Outside </font>

```python
df2_2=df2[df2['Counts']<29]
```

```python
plt.figure(figsize=(12,6))
sns.histplot(df2_2['Counts'],bins=20)
plt.xticks(rotation=90);
print("The mean of the counts is {}.".format(np.mean(df2_2['Counts'])))
print("The sd of the counts is {}.".format(np.std(df2_2['Counts'])))
```

```python
plt.figure(figsize=(20,12))
plt.plot(df2_2['Time'],df2_2['Counts'])
plt.xlabel('Timestamp',size=14)
plt.ylabel('Counts',size=14);
plt.title('Counts vs Time Outdoors',size=24);
```

## Compare the (non-outlier) Data

```python
plt.figure(figsize=(12,6))
sns.distplot(df['Counts'],bins=20,label='Outdoors')
sns.distplot(df2_2['Counts'],bins=20,label='Indoors')
plt.xticks(rotation=90);
plt.legend(loc='upper right');

print("The mean of the outdoors temperature is {}.".format(np.mean(df['Counts'])))
print("The sd of the outdoors temperature is {}.".format(np.std(df2_2['Counts'])))

print("The mean of the indoors temperature is {}.".format(np.mean(df2_2['Counts'])))
print("The sd of the indoors temperature is {}.".format(np.std(df2_2['Counts'])))

print("It seems like the distributions for indoors and outdoors counts overlap.")
np.mean(df['Counts'])-np.mean(df2_2['Counts'])
print("The outdoors counts per minute was on average 0.35 higher than the indoors counts per minute.")
```
